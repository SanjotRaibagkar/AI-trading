```python
import numpy as np
import pandas as pd
import datetime
import random 
import glob
from scipy.signal import find_peaks

import matplotlib.dates as mdates
import matplotlib.pyplot as plt
from mpl_finance import candlestick_ohlc
from matplotlib.dates import MO,AutoDateFormatter,DayLocator,WeekdayLocator
import matplotlib.ticker as ticker

%matplotlib inline
```

```python
def read_data(file):
    price_data=pd.read_csv(file,index_col=0,parse_dates=True,
                          usecols=[u'Date',u'Open Price',u'High Price',u'Low Price',u'Close Price'],
                          dtype={u'Open Price':np.float64,u'High Price':np.float64,
                                u'Low Price':np.float64,u'Close Price':np.float64}).sort_index()
    return(price_data)
```

```python
Stock_name='Balrampurchini'
Stock_price_file='balrampurchini.csv'
data=read_data(Stock_price_file)
sample_data=data.iloc[300:,:].copy()
```

```python
df=pd.read_csv(Stock_price_file,usecols=['Date','Open Price','High Price','Low Price','Close Price'])
df['Date']=pd.to_datetime(df['Date'])
df['Date']=df['Date'].apply(mdates.date2num)
ohlc=df[['Date','Open Price','High Price','Low Price','Close Price']].copy()
f1,ax=plt.subplots(figsize=(16,7))
#print(f1)
candlestick_ohlc(ax,ohlc.values,width=1,colorup='green',colordown='red')
#plt.plot_date(MO,ax,xdate=True)
ax.xaxis_date()
ax.yaxis.tick_right()
ax.yaxis.set_label_position('right')
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title(Stock_name)
plt.rc('axes',grid=True)
plt.rc('grid', color='0.75', linestyle='-', linewidth=0.5)
plt.show()
```
